**Importing Pandas**

```python
import pandas as pd
```

```python
s = pd.Series([10,20,30,40])
print(s)
print(type(s))
```

**Series from a list with Custom Index**

```python
s = pd.Series([85,90,78],index = ["Maths","Science","English"])
print(s)
```

**Series From Tuple**

```python
s = pd.Series((5,10,15))
print(s)
print(type(s))
```

**Series from a Dictionary**

```python
data = {"A":100,"B":200,"C":300}
s = pd.Series(data)
s
```

**Series From Dictionary With Specified Index**

```python
data = {"A":100,"B":200,"C":300}
s = pd.Series(data,index = ["C","B","D"])
s
```

**Series using Scalar Value**

```python
s = pd.Series(5,index = ["a","b","c","d","e"])
s
```

**Series Using numpy array**

```python
import numpy as np
```

```python
arr = np.array([11,22,33])
s = pd.Series(arr)
s
```

**Series With Datatype**

```python
s = pd.Series([1,2,3],dtype="float")
s
```

**Create Empty Series**

```python
s = pd.Series(dtype="int")
s
```

**Series From Range**

```python
s = pd.Series(range(1,6))
s
```

**Series from String Characters**

```python
s = pd.Series(list("DATA"))
s
```

**Series Using Index Only**

```python
s = pd.Series(index=["a","b","c"])
s
```

**Series With Name**

```python
s = pd.Series([10,20,30],name = "Scores")
print(s)
print(s.name)
```

**Pandas cannot store NaN in an integer Series, so it automatically upcasts the whole Series to float64**

```python
s = pd.Series({
    "A":100,"B":200
},index=["A","B","D"])
print(s)
print(s.dtype)
```

```python
s1 = pd.Series([1, 2, 3], index=["a", "b", "c"])
s2 = pd.Series([10, 20, 30], index=["b", "c", "d"])

result = s1 + s2
print(result)
```

```python
s = pd.Series([10, 20, 30, 40, 50], index=["a", "b", "c", "d", "e"])

print(s["b":"d"])
print("---")
print(s[1:3])
```
